### #1.1

Carga el dataset de casas de California que está en la ruta "./data/california_cities.csv". Muestra parte de su contenido y su descripción. Deshazte de las filas con nulos.

```python
df = pd.read_csv("./data/california_cities.csv")
df
```

```python
df.dropna(inplace=True)
df.info()
```

```python
import pandas as pd

# Ordenar el DataFrame por población total en orden descendente
df = df.sort_values(by='population_total', ascending=False)

# Seleccionar las primeras filas para obtener las ciudades más pobladas
ciudades_mas_pobladas = df.head()

# Ahora ciudades_mas_pobladas contendrá las ciudades más pobladas del DataFrame
ciudades_mas_pobladas

# Cambiar Formato a miles en columna "population_total"
ciudades_mas_pobladas['population_total'] = ciudades_mas_pobladas['population_total'] / 1000

ciudades_mas_pobladas
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

PRIMER GRAFICO 

```python

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(10,8))

# Obtener los datos necesarios del DataFrame ciudades_mas_pobladas
ciudades = ciudades_mas_pobladas['city'].tolist()
poblaciones = ciudades_mas_pobladas['population_total'].tolist()

# Colores para las barras
colores = ['blue', 'gray', 'olive', 'brown', 'orange']

# Graficar barras de población con colores específicos
ax.bar(ciudades, poblaciones, color=colores, alpha=0.4)

# Añadir números de población
for i, poblacion in enumerate(poblaciones):
    ax.text(i, poblacion, str(poblacion), ha='center', va='bottom')

# Configuración adicional
ax.set_ylabel('Población')
ax.set_title('Las 5 ciudades más pobladas')


# Mostrar gráfico
plt.show()





```

SEGUNDO GRAFICO 

```python

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(10,8))

# Graficar histograma del área total ocupada por las ciudades
ax.hist(df['area_total_sq_mi'], bins=50, color='steelblue')  # Utilizando el color steelblue

# Configuración adicional
ax.set_xlabel('Área Total')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma del Área Total Ocupada por las Ciudades')

# Cambiar fuente y tamaño del texto
plt.rcParams['font.family'] = 'Bradley Hand ITC'
plt.rcParams['font.size'] = 18

# Mostrar gráfico
plt.show()
```

TERCER GRAFICO 

```python


# Crear figura y ejes
fig, ax = plt.subplots()

# Graficar scatter plot de área de agua vs población
scatter = ax.scatter(df['area_water_sq_mi'], df['population_total'], color='green', alpha=0.5)

# Configuración adicional
ax.set_xlabel('Área de Agua (millas cuadradas)')
ax.set_ylabel('Población')
ax.set_title('Área de Agua vs Población de las Ciudades')

# Mostrar gráfico
plt.show()

```

CUARTO GRAFICO 

```python
# Crear figura y ejes
fig, ax = plt.subplots()

# Graficar boxplot horizontal de la altura de las ciudades
boxplot = ax.boxplot(df['elevation_m'], vert=False, patch_artist=True, boxprops=dict(facecolor='red'))

# Configuración adicional
ax.set_xlabel('Altura (metros)')
ax.set_title('Boxplot de la Altura de las Ciudades')

# Mostrar gráfico
plt.show()
```

FIGURA

```python
# Crear figura y ejes para los 4 subgráficos
fig, axs = plt.subplots(2, 2, figsize=(15, 12))

# Primer gráfico: Las 5 ciudades más pobladas
ciudades = ciudades_mas_pobladas['city'].tolist()
poblaciones = ciudades_mas_pobladas['population_total'].tolist()
colores = ['blue', 'gray', 'olive', 'brown', 'orange']
axs[0, 0].bar(ciudades, poblaciones, color=colores, alpha=0.4)
for i, poblacion in enumerate(poblaciones):
    axs[0, 0].text(i, poblacion, str(poblacion), ha='center', va='bottom')
axs[0, 0].set_ylabel('Población')
axs[0, 0].set_title('Las 5 ciudades más pobladas')

# Segundo gráfico: Histograma del área total ocupada por las ciudades
axs[0, 1].hist(df['area_total_sq_mi'], bins=50, color='steelblue')
axs[0, 1].set_xlabel('Área Total')
axs[0, 1].set_ylabel('Frecuencia')
axs[0, 1].set_title('Histograma del Área Total Ocupada por las Ciudades')

# Tercer gráfico: Scatter plot de área de agua vs población de las ciudades
axs[1, 0].scatter(df['area_water_sq_mi'], df['population_total'], color='green', alpha=0.5)
axs[1, 0].set_xlabel('Área de Agua (millas cuadradas)')
axs[1, 0].set_ylabel('Población')
axs[1, 0].set_title('Área de Agua vs Población de las Ciudades')

# Cuarto gráfico: Boxplot horizontal de la altura de las ciudades
boxplot = axs[1, 1].boxplot(df['elevation_m'], vert=False, patch_artist=True, boxprops=dict(facecolor='red'))
axs[1, 1].set_xlabel('Altura (metros)')
axs[1, 1].set_title('Boxplot de la Altura de las Ciudades')

# Ajustar el espaciado entre subgráficos
plt.tight_layout()

# Mostrar la figura con los 4 gráficos
plt.show()
```
