### #1.1

Carga el dataset de casas de California que est en la ruta "./data/california_cities.csv". Muestra parte de su contenido y su descripcin. Deshazte de las filas con nulos.

```python
df = pd.read_csv("./data/california_cities.csv")
df
```

```python
df.dropna(inplace=True)
df.info()
```

```python
import pandas as pd

# Ordenar el DataFrame por poblacin total en orden descendente
df = df.sort_values(by='population_total', ascending=False)

# Seleccionar las primeras filas para obtener las ciudades ms pobladas
ciudades_mas_pobladas = df.head()

# Ahora ciudades_mas_pobladas contendr las ciudades ms pobladas del DataFrame
ciudades_mas_pobladas

# Cambiar Formato a miles en columna "population_total"
ciudades_mas_pobladas['population_total'] = ciudades_mas_pobladas['population_total'] / 1000

ciudades_mas_pobladas
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

PRIMER GRAFICO 

```python

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(10,8))

# Obtener los datos necesarios del DataFrame ciudades_mas_pobladas
ciudades = ciudades_mas_pobladas['city'].tolist()
poblaciones = ciudades_mas_pobladas['population_total'].tolist()

# Colores para las barras
colores = ['blue', 'gray', 'olive', 'brown', 'orange']

# Graficar barras de poblacin con colores especficos
ax.bar(ciudades, poblaciones, color=colores, alpha=0.4)

# Aadir nmeros de poblacin
for i, poblacion in enumerate(poblaciones):
    ax.text(i, poblacion, str(poblacion), ha='center', va='bottom')

# Configuracin adicional
ax.set_ylabel('Poblacin')
ax.set_title('Las 5 ciudades ms pobladas')


# Mostrar grfico
plt.show()





```

SEGUNDO GRAFICO 

```python

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(10,8))

# Graficar histograma del rea total ocupada por las ciudades
ax.hist(df['area_total_sq_mi'], bins=50, color='steelblue')  # Utilizando el color steelblue

# Configuracin adicional
ax.set_xlabel('rea Total')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma del rea Total Ocupada por las Ciudades')

# Cambiar fuente y tamao del texto
plt.rcParams['font.family'] = 'Bradley Hand ITC'
plt.rcParams['font.size'] = 18

# Mostrar grfico
plt.show()
```

TERCER GRAFICO 

```python


# Crear figura y ejes
fig, ax = plt.subplots()

# Graficar scatter plot de rea de agua vs poblacin
scatter = ax.scatter(df['area_water_sq_mi'], df['population_total'], color='green', alpha=0.5)

# Configuracin adicional
ax.set_xlabel('rea de Agua (millas cuadradas)')
ax.set_ylabel('Poblacin')
ax.set_title('rea de Agua vs Poblacin de las Ciudades')

# Mostrar grfico
plt.show()

```

CUARTO GRAFICO 

```python
# Crear figura y ejes
fig, ax = plt.subplots()

# Graficar boxplot horizontal de la altura de las ciudades
boxplot = ax.boxplot(df['elevation_m'], vert=False, patch_artist=True, boxprops=dict(facecolor='red'))

# Configuracin adicional
ax.set_xlabel('Altura (metros)')
ax.set_title('Boxplot de la Altura de las Ciudades')

# Mostrar grfico
plt.show()
```

FIGURA

```python
# Crear figura y ejes para los 4 subgrficos
fig, axs = plt.subplots(2, 2, figsize=(15, 12))

# Primer grfico: Las 5 ciudades ms pobladas
ciudades = ciudades_mas_pobladas['city'].tolist()
poblaciones = ciudades_mas_pobladas['population_total'].tolist()
colores = ['blue', 'gray', 'olive', 'brown', 'orange']
axs[0, 0].bar(ciudades, poblaciones, color=colores, alpha=0.4)
for i, poblacion in enumerate(poblaciones):
    axs[0, 0].text(i, poblacion, str(poblacion), ha='center', va='bottom')
axs[0, 0].set_ylabel('Poblacin')
axs[0, 0].set_title('Las 5 ciudades ms pobladas')

# Segundo grfico: Histograma del rea total ocupada por las ciudades
axs[0, 1].hist(df['area_total_sq_mi'], bins=50, color='steelblue')
axs[0, 1].set_xlabel('rea Total')
axs[0, 1].set_ylabel('Frecuencia')
axs[0, 1].set_title('Histograma del rea Total Ocupada por las Ciudades')

# Tercer grfico: Scatter plot de rea de agua vs poblacin de las ciudades
axs[1, 0].scatter(df['area_water_sq_mi'], df['population_total'], color='green', alpha=0.5)
axs[1, 0].set_xlabel('rea de Agua (millas cuadradas)')
axs[1, 0].set_ylabel('Poblacin')
axs[1, 0].set_title('rea de Agua vs Poblacin de las Ciudades')

# Cuarto grfico: Boxplot horizontal de la altura de las ciudades
boxplot = axs[1, 1].boxplot(df['elevation_m'], vert=False, patch_artist=True, boxprops=dict(facecolor='red'))
axs[1, 1].set_xlabel('Altura (metros)')
axs[1, 1].set_title('Boxplot de la Altura de las Ciudades')

# Ajustar el espaciado entre subgrficos
plt.tight_layout()

# Mostrar la figura con los 4 grficos
plt.show()
```
